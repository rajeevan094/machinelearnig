```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (tab separated)
df = pd.read_csv("/content/marketing_campaign.csv", sep='	')

# Preview data
df.head()
```

```python
# Shape of dataset
df.shape

# Column names
df.columns

# Data types and non-null counts
df.info()

# Statistical summary
df.describe()
```

```python
# Count null values
df.isnull().sum()
```

```python
(df.isnull().sum() / len(df)) * 100
```

```python
from datetime import datetime

current_year = datetime.now().year
df['Age'] = current_year - df['Year_Birth']

df[['Year_Birth','Age']].head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Age'])
plt.title("Age Boxplot")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Income'], bins=30, kde=True)
plt.title("Income Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Income'])
plt.title("Income Boxplot")
plt.show()
```

```python
spending_cols = ['MntWines','MntMeatProducts','MntFishProducts',
                 'MntSweetProducts','MntGoldProds']

df['Total_Spending'] = df[spending_cols].sum(axis=1)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Education', y='Total_Spending', data=df)
plt.xticks(rotation=45)
plt.title("Average Spending by Education Level")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Income', y='Total_Spending', data=df)
plt.title("Income vs Total Spending")
plt.show()
```
